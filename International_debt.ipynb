```python
afterclean_main_df = main_df.iloc[:-5]
print("Before column row",main_df.shape)
print("After column row",afterclean_main_df.shape)
```

```python
last_8 = afterclean_main_df.iloc[:,-8:]
b=last_8.isnull().any(axis=1).sum()/len(last_8)*100
print(f'{round(b)}%')
print("Before column column",afterclean_main_df.shape)
afterclean_main_df= afterclean_main_df.drop(columns=afterclean_main_df.columns[-8:])
print("After column column",afterclean_main_df.shape)
afterclean_main_df.drop_duplicates(inplace=True)
```

```python
import numpy as np
afterclean_main_df = afterclean_main_df.replace(["", " ", "..", "NA", "N/A"], np.nan)
afterclean_main_df.fillna(0, inplace=True)
print(afterclean_main_df.isnull().sum())
```

```python
afterclean_main_df.columns = afterclean_main_df.columns.str.strip()
print(afterclean_main_df.columns)
print(afterclean_main_df.shape)
```

```python
filtered_df = afterclean_main_df[["Country Name","Country Code" ,"Series Name",'2000', '2001',
       '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010',
       '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019',
       '2020', '2021', '2022', '2023', '2024']]
print(filtered_df.columns)
```

```python

#Analyze country-wise debt distribution

df = filtered_df.copy()
year_cols = [col for col in df.columns if col.isdigit()]

df[year_cols] = df[year_cols].apply(pd.to_numeric)
df

#year wise debt-2006
print("Year-2006 debt of all countries\n",df.groupby('Country Name')['2007'].sum())

#year wise debt
country_data = df.groupby('Country Name')[year_cols].sum()
country_data

#Total Debt
country_data['Total_Debt'] = country_data.sum(axis=1)
print(country_data[['Total_Debt']])
```

```python
country_sorted = country_data.sort_values(by='Total_Debt', ascending=False)


print("Highest debt\n",country_sorted[['Total_Debt']].head(10))

print("\nlowest debt\n",country_sorted[['Total_Debt']].tail(10))

```

```python
analysis_df = filtered_df[
    filtered_df['Series Name'].str.contains('debt per capita', case=False)
].copy()     

year_cols = [col for col in analysis_df.columns if col.isdigit()]

analysis_df['Avg_Debt_Per_Capita'] = analysis_df[year_cols].mean(axis=1)
country_rank = analysis_df[['Country Name', 'Avg_Debt_Per_Capita']]\
    .sort_values(by='Avg_Debt_Per_Capita', ascending=False)

country_rank.head(10)
analysis_df
```

```python
# Total across all years (per country)
analysis_df['Total_Debt_Per_Capita'] = analysis_df[year_cols].sum(axis=1)

country_dist = analysis_df[['Country Name', 'Total_Debt_Per_Capita']]\
    .sort_values(by='Total_Debt_Per_Capita', ascending=False)

# Percentage
total = country_dist['Total_Debt_Per_Capita'].sum()
country_dist['Percentage'] = (country_dist['Total_Debt_Per_Capita'] / total) * 100

print(country_dist.head(10))
```
